```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df
```

```python
#replacing blanks in total charges with 0
```

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
# convert 0 and 1 in citizen citizen columt to yes or no for easy understanding
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("count of churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn' : "count"})
plt.pie(gb['Churn'], labels = gb.index, autopct ="%1.2f%%" )
plt.title("%age of churn")
plt.show()
```

```python
#we can see 26.5% cuuustomers churned
# now let's know why.
```

```python

ax = sns.countplot(x = "gender", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by gender")
plt.show()
```

```python

ax = sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churn by Senior Citizens")
plt.show()
```

```python
#More rate of churn in senior citizens.
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x = "tenure", data = df, bins = 72,hue = "Churn")
plt.show()
```

```python
#People who have used services for long time have stayed. Churned has been done between 1-2 months
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue = "Churn")
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churn by contract")
plt.show()
```

```python
#People with month to month churn more than people with 1 0r 2 year contract
```

```python

# List of categorical columns
cols = [
    "PhoneService", "MultipleLines", "InternetService", "OnlineSecurity",
    "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"
]

# Create subplots â€” adjust rows and cols as you like
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # 3x3 grid for 9 plots
axes = axes.flatten()  # flatten to easily iterate

for i, col in enumerate(cols):
    ax = axes[i]
    sns.countplot(x=col, data=df, hue="Churn", ax=ax)
    for container in ax.containers:
        ax.bar_label(container, fontsize=8)
    ax.set_title(f"Churn by {col}", fontsize=10)
    ax.set_xlabel("")
    ax.set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Overview:
# The following subplots show customer churn distribution across multiple service features.
# Key observations:
# - Churn is higher among customers with Fiber Optic internet.
# - Customers with OnlineSecurity, OnlineBackup, TechSupport, or DeviceProtection tend to churn less.
# - Streaming services (TV, Movies) show little difference in churn behavior.
# - Additional support and protection services appear to improve customer retention.
```
